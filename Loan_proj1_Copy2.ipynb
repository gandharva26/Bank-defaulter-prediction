```python
import pandas as pd
import numpy as np
```

```python
loan=pd.read_excel("Document from Gandharva.xlsx")
```

```python
loan.ndim
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
LevelEncod=LabelEncoder()
loan["grade"]=LevelEncod.fit_transform(loan["grade"])
loan["term"]=LevelEncod.fit_transform(loan["term"])
loan["pub_rec_bankruptcies"]=LevelEncod.fit_transform(loan["pub_rec_bankruptcies"])
loan.head()
loan.loc[loan.loan_status=='Fully Paid','loan_status']=1
loan.loc[loan.loan_status=='Charged Off','loan_status']=0
loan.loc[loan.loan_status=='Current','loan_status']=1
loan.head()
```

```python
loan['pub_rec_bankruptcies'].value_counts()
```

```python
x=loan.drop(loan[loan.loan_status=='Charged Off'].index)//x data frame for non defaulters
```

```python
x["loan_status"]
```

```python
df3=loan[["installment","int_rate","loan_amnt","max_bal_bc","revol_bal","loan_status","grade","term","pub_rec_bankruptcies"]]

df3.hea
```

```python
df3.head()
y=df3["loan_status"]
```

```python
df3.drop('loan_status',axis=1,inplace=True)
```

```python
df3.drop('max_bal_bc',axis=1,inplace=True)
```

```python
x=df3
x.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
score = logmodel.score(x_test, y_test)
print(score)
```
